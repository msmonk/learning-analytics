#讀入CSV檔

```python
import pandas as pd

df = pd.read_csv('tzdata15.csv') #讀入csv檔
temp=df[['欄位名稱','數值']] #只顯示所需欄位
type=temp.groupby(['欄位名稱'])['數值'].sum().reset_index() #欄位名稱欄中，將相同支出項目(如設備費、業務費)之金額加總
type['year']=2015 #新增年份欄位
print(type)

df16 = pd.read_csv('tzdata16.csv')
temp16=df[['欄位名稱','數值']]
type16=temp16.groupby(['欄位名稱'])['數值'].sum().reset_index()
type16['year']=2016
print(type16)

df17 = pd.read_csv('tzdata17.csv')
temp17=df[['欄位名稱','數值']]
type17=temp17.groupby(['欄位名稱'])['數值'].sum().reset_index()
type17['year']=2017
print(type17)

df18 = pd.read_csv('tzdata18.csv')
temp18=df[['欄位名稱','數值']]
type18=temp18.groupby(['欄位名稱'])['數值'].sum().reset_index()
type18['year']=2018
print(type18)

frames=[type,type16,type17,type18]
result=pd.concat(frames)
print(result)
```
